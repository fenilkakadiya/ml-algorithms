```python
X = df[['Weight(Pounds)']]
Y = df['Height(Inches)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train,Y_train)
```

```python
print("coefficient or slop",regression.coef_)
```

```python
print("intercept",regression.intercept_)
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,regression.predict(X_train))
```

```python
Y_Pred = regression.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse = mean_squared_error(Y_test,Y_Pred)
mae = mean_absolute_error(Y_test,Y_Pred)
rmse = np.sqrt(mse)
```

```python
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(Y_test,Y_Pred)
```

```python
print(score)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = np.array(["Reliability", "Performance", "Usability", "Security"])
current = np.array([75, 65, 85, 70])
target = np.array([95, 85, 95, 90])

# Close the loop
current_loop = np.append(current, current[0])
target_loop = np.append(target, target[0])
labels_loop = np.append(labels, labels[0])

angles = np.linspace(0, 2 * np.pi, len(labels_loop))

# Plot radar chart
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)

ax.plot(angles, current_loop)
ax.fill(angles, current_loop, alpha=0.2)

ax.plot(angles, target_loop)
ax.fill(angles, target_loop, alpha=0.2)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

ax.set_title("Spider-Web Diagram: Air Quality Monitoring System", pad=20)

# Save file
path = "/mnt/data/spider_chart_air_quality.png"
plt.savefig(path, dpi=300, bbox_inches="tight")

path
```
